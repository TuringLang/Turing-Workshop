# Julia, Turing.jl, and SSMProblems: Make them yours

This last part of the tutorial is simple: Take what you've learned today and try to use it to implement something relevant for your own work or interests.

* Do you work with some particular sort of time series models? Maybe try implementing a simple toy example of one using Julia and Turing.
* Don't need Bayesian inference but want to see if Julia would be good for some descriptive stats work? Take a simple dataset you know already and see what you can do with it in Julia.
* Just generally want to familiarise yourself with with Turing.jl and/or SSMProblems? Come up with a toy problem, or maybe take one from [our tutorials](https://turinglang.org/docs/tutorials/coin-flipping/index.html), and mess around with it.
* Have fun.

We'll be around to answer questions and help you out.

In the future, when we are not in the room, if you're working on something Julia or Turing related and want help, the best places are:
* https://discourse.julialang.org/ Any Julia questions. Use the tag `#turing` for Turing-specific questions.
* https://julialang.org/slack/ Likewise, any Julia questions. Join the channel `#turing` to ask Turing-specific questions, and to see announcements of what we are up to.
* https://github.com/TuringLang/Turing.jl/issues If you run into any issues using Turing. They can be bugs, or just something that's confusing and hard to do.
* https://github.com/TuringLang/SSMProblems.jl/issues Likewise for SSMProblems.
